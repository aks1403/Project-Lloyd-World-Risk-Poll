```python
train_errors = []
test_errors = []

for i in range(1,4):
    poly_converter = PolynomialFeatures(degree = i,include_bias = False)
    poly_features = poly_converter.fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.33, random_state=42)
    
    model.fit(X_train,y_train)
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    train_error = np.sqrt(mean_squared_error(y_train,y_train_pred))
    test_error = np.sqrt(mean_squared_error(y_test,y_test_pred))
    
    train_errors.append(train_error)
    test_errors.append(test_error)
    
```

```python
plt.figure(dpi = 200)
plt.plot(range(1,3),train_errors[1:3],label = 'train error')
plt.plot(range(1,3),test_errors[1:3], label = 'test error')
plt.xlabel('Degree of Polynomial Regression')
plt.ylabel('Errors')
plt.legend()
plt.show()
```

```python
plt.figure(dpi = 200)
plt.plot(range(1,4),train_errors[0:3],label = 'train error')
plt.plot(range(1,4),test_errors[0:3], label = 'test error')
plt.xlabel('Degree of Polynomial Regression')
plt.ylabel('Errors')
plt.legend()
plt.show()
```

```python
import statsmodels.api as sm
from scipy.stats import f
```

```python

model_ols = sm.OLS(y_train, X_train).fit()
ols_predictions = model_ols.predict(X_test)

# Fit a polynomial regression model
degree = 2  # You can change the degree to test different polynomial degrees
poly_features = PolynomialFeatures(degree=degree, include_bias = False)
X_poly = poly_features.fit_transform(X)
X_train_poly, X_test_poly, y_train_poly, y_test_poly = train_test_split(X_poly, y, test_size=0.33, random_state=42)
model_poly = LinearRegression().fit(X_train_poly, y_train_poly)
poly_predictions = model_poly.predict(X_test_poly)

# Calculate the Mean Squared Error (MSE) for both models
mse_ols = r2_score(y_test, ols_predictions)
mse_poly = r2_score(y_test_poly, poly_predictions)

# Calculate the degrees of freedom for the F-test
n = len(X_test)
k1 = 2  # Number of parameters for OLS (intercept and slope)
k2 = degree + 1  # Number of parameters for polynomial regression

# Calculate the F-statistic and p-value
F_statistic = ((mse_ols - mse_poly) / (k2 - k1)) / (mse_poly / (n - k2))
p_value = 1 - f.cdf(F_statistic, k2 - k1, n - k2)

# Set the significance level for the test (e.g., 0.05)
alpha = 0.05

# Check if the F-statistic is statistically significant
if p_value < alpha:
    print(f"The polynomial regression model is better (p-value: {p_value:.4f}).")
else:
    print(f"The OLS linear regression model is better (p-value: {p_value:.4f}).")
```
